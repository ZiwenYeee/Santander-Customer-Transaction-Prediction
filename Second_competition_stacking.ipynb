```python
def calculate_test(data,var):
    pop = pd.DataFrame()
    for i in range(0,100):
        low = np.percentile(data[var], i * 1)
        high = np.percentile(data[var], (i + 1) * 1 )
        pop.loc[i, 'low'] = low
        pop.loc[i, 'high'] = high
        pop.loc[i, 'bad'] = data.loc[ (data[var] >= low) & (data[var]< high) &( data.target == 1),'target'].count()
        pop.loc[i, 'good'] = data.loc[ (data[var] >= low) & (data[var] < high) &( data.target == 0),'target'].count()
    return pop
```
