```python
import numpy as np
import pandas as pd
import gc
import time
import lightgbm as lgb
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.model_selection import KFold, StratifiedKFold
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

def display_importances(feature_importance_df_):
    cols = feature_importance_df_[
        ["feature", "importance"]].groupby("feature").mean().sort_values(by="importance", ascending=False)[:100].index
    best_features = feature_importance_df_.loc[feature_importance_df_.feature.isin(cols)]
    plt.figure(figsize=(40, 40))
    sns.barplot(x="importance", y="feature", data=best_features.sort_values(by="importance", ascending=False))
    plt.title('LightGBM Features (avg over folds)')
    plt.tight_layout
    # plt.savefig('D:\\Kaggle\\MyFirstKaggleCompetition\\Data\\lgbm_importances01.png')

def kfold_lightgbm(x_train,x_test, feature, params,num_folds, stratified = False):
    if stratified:
        folds = StratifiedKFold(n_splits= num_folds, shuffle=True, random_state=2)
    else:
        folds = KFold(n_splits= num_folds, shuffle=True, random_state=2)
    ntrain = x_train.shape[0]
    ntest = x_test.shape[0]
    oof_train = np.zeros((ntrain,))
    oof_test = np.zeros((ntest,))
    oof_test_skf = np.empty((num_folds, ntest))
    feature_importance_df = pd.DataFrame()
    for n_fold, (train_idx, test_idx) in enumerate(folds.split(x_train[feature],x_train['target'])):
#         X_train, X_test = x_train[feature].iloc[train_idx],x_train[feature].iloc[test_idx]
#         y_train, y_test = x_train['target'].iloc[train_idx],x_train['target'].iloc[test_idx]
        dtrain = lgb.Dataset(data = x_train[feature].iloc[train_idx],
                            label = x_train['target'].iloc[train_idx],
                            free_raw_data = False, silent = True)
        dtest = lgb.Dataset(data = x_train[feature].iloc[test_idx],
                           label = x_train['target'].iloc[test_idx],
                           free_raw_data = False, silent = True)

        clf = lgb.train(
            params=params,
            train_set=dtrain,
            num_boost_round=100000,
            valid_sets=[dtrain, dtest],
            early_stopping_rounds=400,
            verbose_eval=4000
        )
        oof_train[test_idx] = clf.predict(dtest.data)
        oof_test_skf[n_fold,:] = clf.predict(x_test[feature])
        fold_importance_df = pd.DataFrame()
        fold_importance_df["feature"] = feature
        fold_importance_df["importance"] = clf.feature_importance(importance_type='gain')
        fold_importance_df["fold"] = n_fold + 1
        feature_importance_df = pd.concat([feature_importance_df, fold_importance_df], axis=0)
        print('Fold %2d AUC : %.6f' % (n_fold + 1, roc_auc_score(dtest.label, oof_train[test_idx])))
        del clf, dtrain, dtest, fold_importance_df
        gc.collect()
    print("Full AUC score %.6f" % roc_auc_score(x_train['target'], oof_train))
    oof_test[:] = oof_test_skf.mean(axis=0)
    display_importances(feature_importance_df)
    return oof_train, oof_test, feature_importance_df
```
